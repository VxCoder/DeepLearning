```python
x = tf.range(25) + 1
```

```python
x = tf.reshape(x, [1, 5, 5, 1])
x = tf.cast(x, tf.float32)
```

```python
w = tf.constant([[-1,2,-3.], [4, -5, 6], [-7,8,-9]])
w = tf.expand_dims(w, axis=2)
w = tf.expand_dims(w, axis=3)
w.shape
```

```python
out = tf.nn.conv2d(x, w, strides=2, padding='VALID')
out
```
