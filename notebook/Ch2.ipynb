```python
import numpy as np
from matplotlib import pyplot
from matplotlib import cm
from scipy.misc import derivative
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
def f(x):
    return x**2 * np.sin(x)
```

```python
pyplot.plot(x, f(x))
pyplot.plot(x, derivative(f, x, 1e-6),  linestyle='dashed')
```

```python
x = np.linspace(-1, 1, 200)
y = np.linspace(-1, 1, 200)
def z(x, y):
    return -(np.cos(x)**2 + np.cos(y)**2)**2
```

```python
X, Y = np.meshgrid(x, y)
ax = pyplot.gca(projection='3d')
ax.plot_surface(X, Y, z(X,Y), linewidth=0.2, cmap=cm.coolwarm, antialiased=True)
```
